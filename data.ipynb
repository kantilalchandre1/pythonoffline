```python
#4. Records with an incorrectly formated FEIN 
def check_FEIN_format(FEIN):
    return FEIN.isdigit() and len(FEIN) == 9

incorrect_fein = df[~df['Broker Agency FEIN'].apply(check_FEIN_format)].head(5)
incorrect_fein['Test Scenario'] = 'incorrect_fein'
```
